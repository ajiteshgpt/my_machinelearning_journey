```python
import pandas as pd
import numpy as np
```

```python
house=pd.read_csv('/content/Melbourne_house.csv')
```

```python
house
```

```python
# Drop the null values row from the dataset. (Hints- use dropna method)
house.isnull().sum()
house.dropna(subset='Price',inplace=True)
```

```python
house.info()
```

```python
house.isnull().sum()
house.drop(columns=['Unnamed: 13', 'Unnamed: 14'],inplace=True)
```

```python
house.head()
```

```python
#Print the unique value from “SellerG” column.
house['SellerG'].unique()

#give unique values
```

```python
# Print the value count of the region name.
house['Regionname'].value_counts()
```

```python
# How many unique value in postcode column?
house['Postcode'].nunique()   #give numper of unique values
```

```python
#Number of house in each Suburban value?
house['Suburban'].value_counts()
```

```python
#Which Suburban area 3 rooms is maximum?
house[house["Rooms"]==3].value_counts()
#or
house[house["Rooms"]<=3].value_counts()
```

```python
# Identify the address of each region's maximum-price house.
region = house.loc[house.groupby('Regionname')['Price'].max()]
(region[['Regionname', 'Address', 'Price']])
```

```python
# What is the minimum, maximum, and average price of 2 rooms house?
house[house['Rooms']==2]['Price'].sum()
house[house['Rooms']==2]['Price'].mean()
house[house['Rooms']==2]['Price'].max()
house[house['Rooms']==2]['Price'].min()
```

```python
#Convert the price column value to float.
house['Price'].astype(float)
house
```

#ask piviot table methods

```python
# Create a pivot table of the region name and council area and count the number of
 #houses are present.
pd.pivot_table(house, index='Regionname', columns='CouncilArea', aggfunc='count')
```

```python
#Create a pivot table of the region name and postcode and count the number of houses present.
pd.pivot_table(house, index='Regionname', columns='Postcode', aggfunc='count')
```

```python
# Identify the min, max, and average of distance columns.
house['Distance'].min()
house['Distance'].max()
house['Distance'].mean().astype(int)
```

```python
house.head()
```

```python
# Which council area has the lowest number of 3 rooms house?
house[house['Rooms']==3]['CouncilArea'].value_counts(ascending=True).head(1)
```

```python
# Create a pie chart to visualize the proportion of each property type.
import matplotlib.pyplot as plt
house['Type'].value_counts().plot(kind='pie',autopct='%1.3f%%')
```

```python
# Create a scatter plot to visualize the relationship between distance and price.
plt.figure(figsize=(10,6))
plt.scatter(house['Distance'],house['Price'])
```

```python
#Create a bar chart to compare each property type's average number of rooms.
plt.figure(figsize=(10,6))
plt.bar(house['Type'],house["Rooms"].mean())
```

```python
#Create a histogram chart of price columns.
house['Price'].plot(kind='hist')
```

```python
#Create the line chart for the date and price column.
plt.figure(figsize=(10,2))
house['Date']['Price']
```
