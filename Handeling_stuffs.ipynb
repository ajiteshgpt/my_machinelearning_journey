#comple case analysis (cca) -- removing the missing values

Complete case analysis also called list wise deletion of cases consist in discarding observation(rows) where value in any of variables(columns) are missing


Complete case analysis means literally analysis one of those observations for which there is information in all of variable in data set

---
*Assumption*--not more then 5%

1) date is missing completely at random-- tabhi use karna jab randomply distributed ho

agar kisi parten mai hoga toh ham tab remove nahi karege

---
*Advantages*

1) Easy to implement as no data manupilation required

2) Reserves variable distribution( if data is MCAR )then the distribution of the variable of the reduced data set should match the distribution in the original data set



*disadvantages *

1) It can exclude a large fraction of original data set if missing data is abundant

2) Excluded observation could be inform....ysis if data is not missing at random

3) When using our model is production the model will not know how to handle missing value

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split , GridSearchCV

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv("/content/train.csv")
```

```python
df.head()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace =True)
```

```python
df.head()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(df.drop(['Survived'],axis=1),df['Survived'],test_size=0.2,random_state=2)
```

```python
df.isnull,sum()
```

```python
numerical_features=['Age','Fare']

numerical_trf=Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='median')),
    ('standard',StandardScaler())

])

catagorical_features=['Sex','Embarked']
catagorical_trf=Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='most_frequent')),
    ('ohe',OneHotEncoder(handle_unknown='ignore'))
])
```

```python
preprocessor=ColumnTransformer(transformers=[
    ('mum',numerical_trf,numerical_features),
    ('cat',catagorical_trf,catagorical_features)
])
```

```python
clf=Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('classifier',LogisticRegression())
])
```

```python
from sklearn import set_config
set_config(display='diagram')
clf
```

```python
param_grid = {
    'preprocessor__num__imputer__strategy': ['mean', 'median'],
    'preprocessor__cat__imputer__strategy': ['most_frequent', 'constant'],
    'classifier__C': [0.1, 1.0, 10, 100]
}

grid_search = GridSearchCV(clf, param_grid, cv=10)
```

```python
param_grid = {
    'preprocessor__mum__imputer__strategy': ['mean', 'median'],
    'preprocessor__cat__imputer__strategy': ['most_frequent', 'constant'],
    'classifier__C': [0.1, 1.0, 10, 100]
}

grid_search = GridSearchCV(clf, param_grid, cv=10)
grid_search.fit(X_train, y_train)

print(f"Best params:")
print(grid_search.best_params_)
```

```python
grid_search.best_score_
```
