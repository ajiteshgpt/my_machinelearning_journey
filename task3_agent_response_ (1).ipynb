```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from scipy.sparse import hstack
```

```python
# Vectorize the Text (Convert words to numbers)
# stop_words

tf1 = TfidfVectorizer(stop_words='english',max_features=5000)
X_text_matrics= tf1.fit_transform(df['Full_text'])
```

```python
# rating vector cloumn no need to vectprize already vectorized

X_rating_matrix = df['User_Rating'].values.reshape(-1, 1)  # shapping needed to align with text matrics
```

```python
# main objective is to combine text and rating matrix so machine can understand it easily
#It learns that "Bad" + Rating 1 is definitely negative, but "Bad" (as in "Not bad") + Rating 5 is actually positive

X_final = hstack([X_text_matrics, X_rating_matrix])
```

```python
# here we need to split the data and into 3 catagories

# one for sentiment
# one for problem solved or not
# one for target aspect

X_train, X_test, y_train_sentiment, y_test_sentiment = train_test_split(X_final, df['Sentiment'], random_state=42, test_size=0.2)


X_train, X_test, y_train_problem, y_test_problem = train_test_split(X_final, df['Problem_Solved'], random_state=42, test_size=0.2)
X_train, X_test, y_train_target_aspect, y_test_target_aspect = train_test_split(X_final, df['Target_Aspect'], random_state=42, test_size=0.2)
```

```python
lr1 =LogisticRegression(solver='saga',max_iter=1000)
lr2 =LogisticRegression(solver='saga',max_iter=1000)
lr3 =LogisticRegression(solver='saga',max_iter=1000)
```

```python

lr1.fit(X_train, y_train_sentiment)
lr2.fit(X_train, y_train_problem)
lr3.fit(X_train, y_train_target_aspect)
```

```python



y_pred1 = lr1.predict(X_test)
y_pred2 = lr2.predict(X_test)
y_pred3 = lr3.predict(X_test)

# Calculate accuracy using the specific prediction for each target
acc1 = accuracy_score(y_test_sentiment, y_pred1)
acc2 = accuracy_score(y_test_problem, y_pred2)
acc3 = accuracy_score(y_test_target_aspect, y_pred3)

print(f"Accuracy (Sentiment): {acc1}")
print(f"Accuracy (Problem Solved): {acc2}")
print(f"Accuracy (Target Aspect): {acc3}")
```

```python
def inferance1(user_text , agent_text , feedback , rating):
  combined = user_text + "" + agent_text + "" + feedback + "" + str(rating) # Convert rating to string
  text_vector = tf1.transform([combined]) # Use transform instead of TfidfVectorizer
  rating_vector = np.array([[rating]])
  final_input = hstack([text_vector,rating_vector])
  return lr1.predict(final_input) , lr2.predict(final_input), lr3.predict(final_input)
```

```python
#creating a new dataset df_grouped

df_grouped=df.groupby("ID").agg({
    "User_Prompt": lambda x: "".join(x),
    "Agent_Response": lambda x: "".join(x),
    "Feedback_by_User":lambda x:"".join(x),
    "Sentiment":"last",
    "Target_Aspect":"first",
    "Problem_Solved":"last"

})
```

```python
df_grouped["text"]=df_grouped["User_Prompt"]+""+df_grouped["Agent_Response"]+""+df_grouped["Feedback_by_User"]
```

```python
# now for bert we have to assign maping to the sentiments , target aspect and problem solved or not

map_sentiment = {
    'Positive': 0,
    'Happy': 0,
    'Excited':0,
    'Neutral': 1,
    'Curious': 1,
    'Worried': 2,
    'Mixed': 2,
    'Angry': 3,
    'Frustrated':3,
    'Negative':3,
    'Concerned':3
}
```

```python
df["Target_Aspect"].value_counts()
```

```python

map_topic = {
    'Software': 0,
    'Display': 1,
    'Design': 2,
    'Battery': 3,
    'Connectivity': 4,
    'Heating': 5,
    'Hardware': 4,
    'Camera': 4,
    'Performance': 4,
    'Features': 2,
    'Accessories': 2,
    'Audio': 2,
    'Build': 1,
    'Value': 1
}
```

```python

map_solved = {
    'Yes': 0,
    'No': 1,
    'Pending': 1
}

```

```python
df_grouped['Sentiment_Encoded'] = df_grouped['Sentiment'].map(map_sentiment)
df_grouped['Target_Aspect_Encoded'] = df_grouped['Target_Aspect'].map(map_topic)
df_grouped['Problem_Solved_Encoded'] = df_grouped['Problem_Solved'].map(map_solved)
```
